```python
import pandas as pd 
```

```python
print("Reading csv file ")
df=pd.read_csv('labeled_data.txt')
print("\n file reading sucessfully")
```

```python
print("printing top rows ")
df.head()
```

```python
print(" dropping or removing  unnamed column name  ")
df=df.drop('Unnamed: 0', axis=1)

df.head()
```

```python
print("changing tweet column name to comments")
df.rename(columns={'tweet':'comments'},inplace=True)
```

```python
print("changing count column name to total_votes")
df.rename(columns={'count':'Total_votes'},inplace=True)
```

```python
print("printing the info of the file ")
df.info()
```

```python
print("describing the file ")
df.describe()
```

```python
print("Rows and columns in file are ")
df.shape
```

```python
print("printing class and comments cloumn ")
df[['class','comments']]
```

```python

print("Missing Values in Each Column:")
print(df.isnull().sum())


df = df.dropna()

print("\nDataset Shape After Removing Missing Values:")
df.shape
```

```python

print("Total Duplicate Rows:")
print(df.duplicated().sum())

df = df.drop_duplicates()

print("\nDataset Shape After Removing Duplicates:")
df.shape

print("\nClass Distribution:")
print(df['class'].value_counts())


df.to_csv('cleaned_data.csv', index=False)

print("\nCleaned Dataset Saved Successfully!")
```
